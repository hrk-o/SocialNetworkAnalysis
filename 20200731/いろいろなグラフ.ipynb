```python
# 一次元格子
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

G = nx.Graph()
n = 21 # 頂点の数
G.add_nodes_from([i for i in range(n)])
for i in range(n-1): # 辺の設定
    G.add_edges_from([(i, i+1)])
nx.draw(G, node_size=400, node_color='red', with_labels=True, font_weight='bold')

# グラフ情報
print(nx.info(G)) # 平均次数 : average degree
#plt.bar(len(nx.degree_histogram(G)), height = nx.degree_histogram(G))#次数ヒストグラム

# 平均最短経路長
print("average shortest path length : ",nx.average_shortest_path_length(G)) # average shortest path length

# 直径
print("diameter:" , nx.diameter(G))

#最短経路・経路長
#print(nx.shortest_path(G)[0])
#print(nx.shortest_path_length(G))
```
