```python
import graph2_lib as glib
import matplotlib.pyplot as plt
import japanize_matplotlib # 日本語
import math

lib = glib.graph2_lib()
```

```python
# 完全グラフ(グラフ)

nnode = list() # ノード数
nedge = list() # エッジ数
avedegree = list() # 平均次数
avespl = list() # 平均最短経路長
diameter = list() # 直径
aveclust = list() # クラスタリング係数
checkavespl = list() #平均最短経路長(平均頂点間距離)確認用

for num in range(1,10):
    lib.complete_graph(num)
    nnode.append(lib.getnnode())# ノード数
    nedge.append(lib.getnedge())# エッジ数
    avedegree.append(lib.getavedegree())# 平均次数
    avespl.append(lib.getavespl())# 平均最短経路長
    diameter.append(lib.getdiameter())# 直径
    aveclust.append(lib.getaveclust())# クラスタリング係数
    checkavespl.append(1)# 平均最短経路長(平均頂点間距離)確認用

plt.figure(0)    
plt.plot(nnode, nedge, label="エッジ数")  
plt.plot(nnode, avedegree, label="平均次数")  
plt.plot(nnode, avespl, label="平均最短経路長")  
plt.plot(nnode, diameter, label="直径")  
plt.plot(nnode, aveclust, label="クラスタリング係数") 
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0)

plt.figure(1)
plt.plot(nnode, avespl, label="平均最短経路長")  
plt.plot(nnode, checkavespl, label="平均最短経路長(確認)")  
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0)
```

```python
# 円環（サイクル）（グラフ）

nnode = list() # ノード数
nedge = list() # エッジ数
avedegree = list() # 平均次数
avespl = list() # 平均最短経路長
diameter = list() # 直径
aveclust = list() # クラスタリング係数
checkavespl = list() #平均最短経路長(平均頂点間距離)確認用

for num in range(1,10):
    lib.cycle_graph(num)
    nnode.append(lib.getnnode())# ノード数
    nedge.append(lib.getnedge())# エッジ数
    avedegree.append(lib.getavedegree())# 平均次数
    avespl.append(lib.getavespl())# 平均最短経路長
    diameter.append(lib.getdiameter())# 直径
    aveclust.append(lib.getaveclust())# クラスタリング係数
    checkavespl.append(lib.getnnode() / 4)# 平均最短経路長(平均頂点間距離)確認用

plt.figure(0)    
plt.plot(nnode, nedge, label="エッジ数")  
plt.plot(nnode, avedegree, label="平均次数")  
plt.plot(nnode, avespl, label="平均最短経路長")  
plt.plot(nnode, diameter, label="直径")  
plt.plot(nnode, aveclust, label="クラスタリング係数") 
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0)

plt.figure(1)
plt.plot(nnode, avespl, label="平均最短経路長")  
plt.plot(nnode, checkavespl, label="平均最短経路長(確認)")  
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0)
```

```python
# ツリー（グラフ） d可変、height3

nnode = list() # ノード数
nedge = list() # エッジ数
avedegree = list() # 平均次数
avespl = list() # 平均最短経路長
diameter = list() # 直径
aveclust = list() # クラスタリング係数
checkavespl = list() #平均最短経路長(平均頂点間距離)確認用

for num in range(1,8):
    lib.balanced_tree(num,3)# branching, heightの順。branchingは繋がっているノードの数 heightは階層
    nnode.append(lib.getnnode())# ノード数
    nedge.append(lib.getnedge())# エッジ数
    avedegree.append(lib.getavedegree())# 平均次数
    avespl.append(lib.getavespl())# 平均最短経路長
    diameter.append(lib.getdiameter())# 直径
    aveclust.append(lib.getaveclust())# クラスタリング係数
    checkavespl.append(math.log(lib.getnnode()))# 平均最短経路長(平均頂点間距離)確認用

plt.figure(0)    
plt.plot(nnode, nedge, label="エッジ数")  
plt.plot(nnode, avedegree, label="平均次数")  
plt.plot(nnode, avespl, label="平均最短経路長")  
plt.plot(nnode, diameter, label="直径")  
plt.plot(nnode, aveclust, label="クラスタリング係数") 
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0)

plt.figure(1)
plt.plot(nnode, avespl, label="平均最短経路長")  
plt.plot(nnode, checkavespl, label="平均最短経路長(確認)")  
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0)
```
